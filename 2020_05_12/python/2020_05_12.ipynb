```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import altair as alt
# alt.renderers.enable('notebook')
```

```python
# data
eruptions_ph = 'https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2020/2020-05-12/eruptions.csv'
events_ph = 'https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2020/2020-05-12/events.csv'
sulfur_ph = 'https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2020/2020-05-12/sulfur.csv'
tree_rings_ph = 'https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2020/2020-05-12/tree_rings.csv'
volcano_ph = 'https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2020/2020-05-12/volcano.csv'
```

```python
eruptions = pd.read_csv(eruptions_ph)
events = pd.read_csv(events_ph)
sulfur = pd.read_csv(sulfur_ph)
tree_rings = pd.read_csv(tree_rings_ph)
volcano = pd.read_csv(volcano_ph)
```

```python
# 1 more than 2020 for log 10 scales
volcano['years_ago'] = 2021 - volcano['last_eruption_year']
```

```python
volcano.sort_values('years_ago').head()
```

```python
group_for_plot_log10 = volcano.groupby(['years_ago', 'evidence_category']) \
    .count() \
    .reset_index() 

alt.Chart(group_for_plot_log10).mark_bar(opacity = 0.7).encode(
    x=alt.X('years_ago:O', bin = True),
    y = alt.Y('volcano_number:Q', stack = None),
    color = 'evidence_category')
```

```python
eruptions.columns

group_for_plot_erupt = eruptions.groupby(['start_year', 'eruption_category']) \
    .count() \
    .reset_index() 
group_for_plot_erupt['log_10_start_year'] = np.log10(2021 - group_for_plot_erupt['start_year'])

alt.Chart(group_for_plot_erupt).mark_bar(opacity = 0.7).encode(
    x=alt.X('log_10_start_year:O', bin = True),
    y = alt.Y('volcano_number:Q', stack = None),
    color = 'eruption_category')
```
